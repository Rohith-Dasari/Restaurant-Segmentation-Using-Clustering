```python
#top 5 cuisines
from collections import Counter
final_df['cuisine_list'] = final_df['Cuisines'].apply(lambda x: x.split(', '))
all_cuisines = [cuisine for sublist in final_df['cuisine_list'] for cuisine in sublist]
cuisine_count = Counter(all_cuisines)
top_5_cuisines = cuisine_count.most_common(5)

print("Top 5 Cuisines and their counts:", top_5_cuisines)

```

```python
cuisines, counts = zip(*top_5_cuisines)
plt.pie(counts,labels=cuisines,autopct='%.2f%%')
plt.show()
```
